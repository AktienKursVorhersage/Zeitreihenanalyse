### Imports

```python
#pip install statsmodels
#pip install pmdarima

import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
import statsmodels as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima.arima import auto_arima
from pmdarima.arima.utils import ndiffs

from pandas.plotting import register_matplotlib_converters
```

### Some configurations

```python
register_matplotlib_converters() #Register a datetime converter for a matplotlib plotting method
%matplotlib inline
```

### Load the data

```python
dfDataOrig = pd.read_csv("VOW3DE.csv", #import Stock data of Volkswagen
                     parse_dates = True, #Keep date forma
                     index_col = "Date", #set the Data column as index
                     squeeze = True)

dfData = dfDataOrig.copy() #Create a copy, so the original data is not modified

#print(dfData.head(5))
#print(dfData.tail(5))
#print(dfData.shape) #(5803, 6)
#print(dfData.index) # dtype='datetime64[ns]', name='Date', length=5803, freq=None
```

### Prepare the data

```python
dfData = dfData[["Close"]] # Keep only the Close column

start = dfData.index.searchsorted(pd.datetime(1998,1 , 1))
end = dfData.index.searchsorted(pd.datetime(2020, 1, 1))
dfData = dfData.iloc[start:end] # Keep a specified time frame

dfData = dfData.resample("M").mean() #Resample the data to get the mean for each month in a year
#print(dfData)

#Plot the data
fig = plt.gcf()
fig.set_size_inches(16, 9) # Make the plot bigger

plt.xlabel("Month")
plt.ylabel("Closing price of Volkswagen")
plt.plot(dfData, color="blue")
plt.grid(True)
plt.show()
```

### Check the stationarity 
- Mean and Standard deviation 
- Augumented Dickey fuller test (Null Hypothesis: If p-value > 0.05 than the data is not stationary)

```python
# Function to test the stationarity
def test_stationarity(timeseries):
    #Determing rolling statistics -> Take the previous defined window to calculate the statistics for the next value
    rollmean = timeseries.rolling(12).mean() # 12 because we have 12 Months 
    rollstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    fig = plt.gcf()
    fig.set_size_inches(16, 9)
    
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rollmean, color='red', label='Rolling Mean')
    plt.plot(rollstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.grid(True)
    plt.show(block=False)
    
    print("Results of dickey fuller test")
    adft = adfuller(timeseries, autolag="AIC")
    # output for adft will give us return values without defining what the values are.
    # hence we manually write what the values represent
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)  
```

```python
test_stationarity(dfData)
# p-value 0.685343 -> Bigger than 0.05, so the Null Hypothesis cannot be rejected -> Data is not stationary
```

```python
dfData_Diff1 = dfData.diff(periods=1).dropna() # Take the difference of a month with the next month
#dfData_Diff1 = dfData.diff(periods=1).fillna(dfData) # Take the difference of a month with the next month

test_stationarity(dfData_Diff1) #p-value 2.450645e-19 -> Smaller than 0.5, so we can reject the Null Hyptothesis -> Data is stationary

auto_ndiffs = ndiffs(dfData, test="adf") # Check with library, which checks automatically the order of differancing
```

```python
# Get Auto Regression and Moving Average Value
plot_acf(dfData_Diff1) # Moving Average (MA) term = 1 (q) -> Spike at 1
plt.show()
plot_pacf(dfData_Diff1) # Auto Regressive (AR) term = 1 (p) -> Spike at 1
plt.show()

modelAutoArima = auto_arima
```

### Train the ARIMA model

```python
model_ARIMA = ARIMA(dfData,order=(1,1,1))
fitted_ARIMA = model_ARIMA.fit()
print(fitted_ARIMA.summary())

# Plot arima
fig = plt.gcf()
fig.set_size_inches(16, 9)
    
plt.plot(dfData, color='blue',label='Original')
plt.plot(fitted_ARIMA.fittedvalues, color='red', label='ARIMA Analysis')
plt.legend(loc='best')
plt.title('Original and ARIMA Analysis')
plt.grid(True)
plt.show(block=False)
```

### Predict with ARIMA

```python
stepsARIMA = 24
predict_ARIMA = fitted_ARIMA.forecast(steps=stepsARIMA) # Predict next to two months

#Compare Prediction to Original
dfARIMADataActual = dfDataOrig.copy()
dfARIMADataActual = dfARIMADataActual[["Close"]] 

index_for_arima = pd.date_range(dfData.index[-1], periods = stepsARIMA + 1 , freq='M') #Create Series with the dates for predicited values
index_for_arima = index_for_arima[1:] #remove first entry, because it's the last date in the training data

dfARIMADataActual = dfARIMADataActual[index_for_arima[0]:index_for_arima[-1]] # Keep a specified time frame
dfARIMADataActual = dfARIMADataActual.resample("M").mean() 

print(dfARIMADataActual)

fig = plt.gcf()
fig.set_size_inches(16, 9)
    
plt.plot(dfARIMADataActual, color='blue',label='Original')
plt.plot(predict_ARIMA, color='red', label='ARIMA Prediction')
plt.legend(loc='best')
plt.title('Original and ARIMA Prediction')
plt.grid(True)
plt.show(block=False)
```

```python
smodel = auto_arima(dfData, start_p=0, start_q=0,
                         test='adf',
                         max_p=4, max_q=4, m=12,
                         start_P=0, start_Q=0, seasonal=True,
                         d=auto_ndiffs, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)
```

```python
n_periods = 24

prediction_SARIMA, confint = smodel.predict(n_periods=n_periods , return_conf_int=True, alpha=0.05) #Predicted Values and Confidence Intervall of 95%

index_of_series = pd.date_range(dfData.index[-1], periods = n_periods + 1 , freq='M') #Create Series with the dates for predicited values
index_of_series = index_of_series[1:] #remove first entry, because it's the last date in the training data

dfDataActual = dfDataOrig[index_of_series[0]:index_of_series[-1]] #Get the actual values from the original data
dfDataActual = dfDataActual[["Close"]]
dfDataActual = dfDataActual.resample("M").mean() 

prediction_series = pd.Series(prediction_SARIMA, index=index_of_series) #Prepare prediction data for plotting
lower_series = pd.Series(confint[:, 0], index=index_of_series) #Prepare lower bound of confidence intervall for plotting
upper_series = pd.Series(confint[:, 1], index=index_of_series) #Prepare upper bound of confidence intervall for plotting

# Plot arima
fig = plt.gcf()
fig.set_size_inches(16, 9)

plt.plot(dfData, color='blue',label='Train data') #Include the train data in the model
plt.plot(dfDataActual, color='green',label='Actual Values')
plt.ylabel("Closing Price Volkswagen")
plt.xlabel("Date in Months")

plt.plot(prediction_series, color='red', label='ARIMA Prediction')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.05)
plt.legend(loc='best')
plt.title('Original vs. ARIMA Predicition')
plt.grid(True)
plt.show(block=False)
```
